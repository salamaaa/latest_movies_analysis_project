```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
median_date = df['release_date'].median()
df['release_date'] = df['release_date'].fillna(median_date)
```

```python
df['overview'] = df['overview'].fillna('Unknown Overview')
```

```python
from sqlalchemy import create_engine
import urllib.parse

#Step1: Connect to PostrgreSQL

username = "postgres"
password = urllib.parse.quote_plus("123Aa@456") #encode special characters
host = "localhost"
port = 5433
database = "latest_2025_movies_db"

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#Step2: Load DataFrame into PostgreSQL
table_name = "movie"
df.to_sql(table_name,engine,if_exists="replace",index=False)
print(f"Data Successfull loaded into table '{table_name}' in database '{database}'")
```
